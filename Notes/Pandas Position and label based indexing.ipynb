### Position and Label based indexing using `Df.iloc` and `Df.loc` functions

-  There are other ways to indexing the dataframe which pandas recomended and they are more explicite and more ambiguous
-  There are two main way to indexing the dataframe.
    -  Position based indexing using `df.iloc`.
    -  Label based indexing using `df.loc`.

```python
import pandas as pd
import numpy as np
market_df = pd.read_csv(r"F:\PGD_UpGrad\Preparatory\Module 5 Python for Data Science\Referance Data\Introduction+to+Pandas\Introduction to Pandas\global_sales_data/market_fact.csv")
market_df.head()
```

### Position(Integer) based indexing:
-  Panda provide the `df.iloc` functionality to index the dataframe using integer indices.

```python
# here the 2 is 3rd row and 4 is 5th column of the market_df dataframe.
market_df.iloc[2,4]
```

-  Note that if we ditectly give the `df[2,4]` will throw an error, since pandas get confused whether 2 is a index or it is row with label = 2?
-  `df.iloc[2,4]` tell pandas explicitly that it should assume integer indices.

```python
# it select the single row with all columns.
market_df.iloc[5]
```

```python
# we can select the single row with all column as given below also
# where 5 is 6th row and : indicate that cloumn form 0th to last 
market_df.iloc[5,:]
```

```python
# we can select the multiple rows using list of indcies
market_df.iloc[[2,3,4]]
```

```python
# WE CAN WRITE ABOVE EXPRESSION AS GIVENabs
market_df.iloc[[2,3,4],:]
```

```python
# selecting row using range
market_df.iloc[4:8]
```

```python
# we can write above expression as
market_df.iloc[4:8,:]
```

```python
# selecting the single column
market_df.iloc[:,2].head()
```

```python
# selcting multiple columns
market_df.iloc[:,3:8].head()
```

```python
# selectin multiple row and columns
market_df.iloc[2:6,0:5]
```

```python
# Using the booleans
# This select the row corrosponding to the True 
```

```python
market_df.iloc[[True, False,True, False, False,False,True],[True,False,True,True,True,True,False]]
```

-  `df.iloc` use integer indices starting from the 0.
-  The other common way to indexing is label based index using `df.loc`.

### Label based indexing: 

-  Pandas provid the `df.loc` for indexing the dataframe using `Label based indexing`

```python
market_df.loc[2,'Sales']
```

```python
market_df.set_index('Ord_id', inplace=True)
market_df.head()
```

```python
# Select single row with some columns
market_df.loc['Ord_5406',['Prod_id','Sales','Discount','Profit']]
```

```python
# select multiple rows and columns
market_df.loc[['Ord_5406','Ord_5456','Ord_5446'],'Prod_id':'Profit']
```
