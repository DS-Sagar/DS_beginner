### Pandas series:
-  It is almost similar to the numpy array. it contains the scalar value of the same type (e.g Numbers, chracters, date etc.)
-  A data frame is a table where each column is a pandas series.

#### Creating a pandas series
-  Series is a array like one-dimetional structure. though unlike the numpy array it contain non-numeric data (charcters, date, time, boolean etc.)
-  We can create the panda series using the `pd.series()`.

```python
import pandas as pd
import numpy as np

series_test = pd.Series([1,2,3,4,5,6,7])
print(series_test)
```

-  Each element of the pandas Series has a index and often it tart from the 0.

```python
#series of the chracter 
char_Series = pd.Series(['a','b','hii','hello'])
print(char_Series)
```

```python
# Series of dates
# Date format should be the Month-day-year
Date_Series = pd.date_range(start = '03-06-2019', end = '03-30-2019')
Date_Series
```

#### Indexing the Pandas Series:

-  Indexing of the series is same as the arrays.

```python
import pandas as pd
import numpy as np

test_series = pd.Series([13.4,12,43,4,63,5,67,45.4])
print(test_series)
```

```python
print(test_series[3])
```

#### Explicitly indexing the series:
-  You may noticed that while creating the series panda is automatically assign the indexing from `0 to (n-1)` which is the row number of data frame but you can give your own indexing aslo while creting the series using `pd.Series`.

```python
pd.Series([12,48,75,48,6,9,5],index = range(10,17))
```

```python
pd.Series(np.array(range(1,10))**2, index = range(11,20))
```
